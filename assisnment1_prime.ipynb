Q1. <span style="color:red;">Prime Number Check using Command Line Argument?</span>

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

try:
    user_input = input("Enter a number: ")

    if user_input.strip() == "1":
        print("You did not enter anything.")
    else:
        num = int(user_input)

        if is_prime(num):
            print(num, "is a Prime Number")
        else:
            print(num, "is NOT a Prime Number")

except ValueError:
    print("Please enter a valid integer number.")
```

Q2. <span style="color:red;">Difference between is and ==?</span>

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print("a == b :", a == b)   # Value comparison
print("a is b :", a is b)   # Memory location comparison

print("a == c :", a == c)
print("a is c :", a is c)

print("\n== compares values, 'is' compares memory location.")
```

Q3. <span style="color:red;">Iterate list, skip negatives, stop at zero, handle exceptions?</span>

```python
# Q3 List Handling

data = [5, -2, 8, "hello", 0, 10]

for item in data:
    try:
        num = int(item)

        if num == 0:
            print("Zero found. Stopping execution.")
            break

        if num < 0:
            continue

        print("Valid number:", num)

    except ValueError:
        print("Non-integer value skipped:", item)
```

Q4. <span style="color:red;">List comprehension (Squares divisible by 2 and 5)?</span>


```python
# Q4 List Comprehension

squares = [x**2 for x in range(1, 21) if x % 2 == 0 and x % 5 == 0]
print("Squares divisible by 2 and 5:", squares)
```

Q5. <span style="color:red;">Find largest and smallest without built-in functions?</span>


```python
# Q5 Largest and Smallest

def find_min_max(lst):
    smallest = lst[0]
    largest = lst[0]

    for num in lst:
        if num < smallest:
            smallest = num
        if num > largest:
            largest = num

    return smallest, largest

numbers = [12, 45, 2, 89, 34]
small, large = find_min_max(numbers)

print("Smallest number:", small)
print("Largest number:", large)
```
